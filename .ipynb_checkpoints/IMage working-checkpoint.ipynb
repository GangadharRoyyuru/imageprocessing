```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('test.jpg',0)
ret,thr1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)            #simple thresholding and its types
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

plt.subplot(231),plt.imshow(img,cmap='gray'),plt.title("Original Image"),plt.axis("off")
plt.subplot(232),plt.imshow(thr1,cmap='gray'),plt.title("BINARY"),plt.axis("off")
plt.subplot(233),plt.imshow(thresh2,cmap='gray'),plt.title("BINARY_INV"),plt.axis("off")
plt.subplot(234),plt.imshow(thresh3,cmap='gray'),plt.title("TRUNC"),plt.axis("off")
plt.subplot(235),plt.imshow(thresh4,cmap='gray'),plt.title("TOZERO"),plt.axis("off")
plt.subplot(236),plt.imshow(thresh5,cmap='gray'),plt.title("TOZERO_INV"),plt.axis("off")
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('thresh1.jpg',0)
ret,th1 = cv2.threshold(img,127,200,cv2.THRESH_BINARY)       #adaptive thresholding
th2 = cv2.adaptiveThreshold(img,200,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,3)
plt.subplot(1,3,1),plt.imshow(img,cmap = 'gray'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(th1,cmap = 'gray'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(th2,cmap = 'gray'),plt.axis('off')
```

```python
import cv2
import numpy as np                     #histogram equalised/
import matplotlib.pyplot as plt

img = cv2.imread('test.jpg',0)
equ = cv2.equalizeHist(img)
plt.figure()
plt.imshow(img,cmap='gray'),plt.axis('off'),plt.title('Original')
plt.figure()
plt.imshow(equ,cmap='gray'),plt.axis('off'),plt.title('Histogram Equalized')

plt.show()
```

```python
hist = cv2.calcHist([img], [0], None, [256],[0,256])
hist2 = cv2.calcHist([equ], [0], None, [256],[0,256])
plt.subplot(2,1,1), plt.plot(hist)
plt.subplot(2,1,2), plt.plot(hist2)
```

```python
img = cv2.imread('test.jpg',1)
kernal = (np.ones((5,5), dtype = float))/25           #image filtering    used for reducing noise, or contrasting edges for details
filimg = cv2.filter2D(img, -1, kernal)
cv2.imshow('scenary filterin', filimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('imtofilter.png',1)
mfilimg = cv2.medianBlur(img,3)                 # median filtering effective for noise reduction
cv2.imshow('median filterin', mfilimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('test.jpg',1)
kernal = np.array([[-1,-1,-1],
                 [-1,9,-1],
                 [-1,-1,-1]])
sharpimg = cv2.filter2D(img,-1, kernal)
cv2.imshow('scenary sharpened', sharpimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("test.jpg",1)
de = (np.ones((3,3), dtype = float))/9                  #sharpening image
ori  = np.zeros((3,3))
ori[2][2] = 1
kernal  = ori - de
detail = cv2.filter2D(img, -1, kernal)

sharpker = ori+kernal                                  #original + detail(edge image) = sharpened image(visible edges)
imgsharpen = cv2.filter2D(img, -1, sharpker)
cv2.imwrite('sharpened.jpg',imgsharpen)
cv2.imshow('scenary sharpened', imgsharpen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#sobel kernal vertical
import cv2
import numpy as np
from matplotlib import pyplot as plt
kernal = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])
img = cv2.imread('catedge.png',0)
ed = cv2.filter2D(img,-1,kernal)
cv2.imshow('edges', ed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#sobel kernal vertical
import cv2
import numpy as np
from matplotlib import pyplot as plt
kernal = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])
ker = np.transpose(kernal)
img = cv2.imread('catedge.png',0)
ed = cv2.filter2D(img,-1,ker)
cv2.imshow('edges', ed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#laplacian kernal edge detection
import cv2
import numpy as np
from matplotlib import pyplot as plt
kernal = np.array([[-1,-1,-1],
                  [-1,8,-1],
                  [-1,-1,-1]])
img = cv2.imread('catedge.png',0)
ed = cv2.filter2D(img,-1,kernal)
cv2.imshow('edges', ed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#erosion
img  = cv2.imread('catedge.png',1)
ker  = np.ones((3,3))
er  = cv2.erode(img,ker,1)
cv2.imshow('erosion', er)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('erosion', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#dialation
img = cv2.imread('catedge.png',0)
dia  = cv2.dilate(img,ker,1)
cv2.imshow('dialation', dia)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#morphological transformations
img = cv2.imread("catedge.png",0)
mor = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,ker)
cv2.imshow('OUTLINE', mor)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
sob = cv2.Sobel(img,-1,1,1,5)
cv2.imshow('sobel', sob)
cv2.waitKey(0)                #sobel
cv2.destroyAllWindows()
```

```python
schar = cv2.Scharr(img,-1,1,0)
cv2.imshow('Scharr', schar)       #scahrr
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lap = cv2.Laplacian(img,-1)
cv2.imshow('Laplace',lap)                   #laplacian
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.GaussianBlur(img,(5,5),0)
can = cv2.Canny(img,20,50,5,L2gradient = True)      #canny image detection
cv2.imshow('canny edge', can)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("chess.jpg",0)                    #hrrris corner detection
harcor = cv2.cornerHarris(img,3,5,0.05)
cv2.imshow('corners', harcor)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("chess.jpg",0)
corners = cv2.goodFeaturesToTrack(img,10,0.1,5)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),5,(0,255,0),-1)
    
plt.subplot(122),plt.imshow(img,cmap='gray'),plt.axis('off'),plt.title('Shi-Tomasi Corners')
plt.show()
```
